# 정규표현식

정규 표현식은 문자열을 일정한 형식에 맞춰 검색, 추출, 수정 등을 하는 방식입니다.  

데이터 분석에서는 해당 문자열에 필요한 문자가 있는지를 확인 후  
이를 추출하는 과정으로 많이 사용합니다.  

또한, 많은 데이터 값을 한번에 수정할 때에도 사용합니다.  

---

## 메타 문자
정규표현식에 사용하는 문자는 다음과 같습니다.  
```
. ^ $ * + ? \ | ( ) { } [ ]  
```

### __반복__에 사용하는 메타문자  
`*` : 0회 이상 반복한 문자열 = `{0,}`  
`+` : 1회 이상 반복한 문자열 = `{1,}`  
`?` : 0회 또는 1회 반복한 문자열 = `{0, 1}`  
`{m}` :  m회 만큼 반복한 문자열  
`{m, n}` :  m회에서 n회까지 반복한 문자열  

__최소__ 몇 회 이상 반복에 사용하는 할 때는   
```
* + 뒤에 ?를 붙여준다.
```
`*?` : 최소 0회 이상 반복한 문자열  
`+?` : 최소 1회 이상 반복한 문자열


### __매칭__에 사용하는 메타문자
`.` : `\n`(줄바꿈)을 제외한 모든 문자와 매치  
`^` : `^문자`, `^` 뒤의 문자와 시작하는 단어와 매치, 여러줄일 경우 한 줄 만 매치  
`$` : `문자$`, `$`앞의 문자로 끝나는 단어와 매치  
`[ ]` : `[문자]`, `[]` 안에 있는 문자 하나와 매치  
`|` : `문자1|문자2`, 문자1 또는 문자2 매치로 or의 의미  
`( )` : `(문자)`, `()` 안에 매치하는 문자를 하나의 그룹으로 묶음   

__문자__그 자체를 매칭할 땐   
__문자__ 그대로를 써준다.   

`[]` 안에서 `^`를 써준다면    
`[]`안에 있는 문자와 일치하지 않는 것을 매치시켜 준다.  
`[^ ]` : `[^문자]`, 문자와 맞지 않는 다른 문자들과 매치

### 이스케이프와 패턴 문자
이스케이프 문자 `\`는  
메타문자를 일반 문자로 취급 시켜준다.  
`\\` : 문자 그대로 `\`
`\d` : 모든 숫자와 매치된다. = `[0-9]`  
`\D` : 숫자가 아닌 모든 문자와 매치된다. = `[^0-9]`  
`\w` : 모든 문자 또는 숫자와 매친된다. = `[A-Za-z0-9]` = `[a-zA-Z0-9]`  
`\W` : 문자가 아닌 모든 문자와 매치된다. = `[^A-Za-z0-9]`    
`\s` : 화이트 스페이스 문자(` `, `\t`, `\n` 등) 매치 된다. = `[ \t\n\r\f\v]`  
`\S` : ` `이 아닌 것들 과 매치된다. = `[^ \t\n\r\f\v]`  
`\b` : 단어의 경계를 나타내며, 단어 문자와 비단어 문자 사이가 매치된다.  
`\B` : 단어 문자와 단어 문자 사이, 비단어 문자와 비단어 문자 사이가 매치된다.  
`\A` : `\A문자`, `\A` 뒤에 문자로 시작하는 문자열과 매치 = `^`
`\Z` : `문자\Z`, `\Z` 앞에 문자로 끝나는 문자열과 매치 = `$`


`\b`와 `\B`
'I love you'라는 문장에서  
love만 일치시키고 lovely와 loved, loves는 매치 시키지 못하게 하는 방법은?  
비문자 + love + 비문자 형식  
`\bov\b`로 표현 해 'love'만 일치시킬 수 있다.

'learn'은 매치 하지만 earn, earning, hear은 일치 시키지 못하는 방법은?  
'l'(문자) + 'ear'(문자) + 'n'(문자) 형식  
`\Bear\B` 로 표현해 'learn'만 일치 시킬 수 있다.
