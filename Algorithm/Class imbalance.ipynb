# Class imbalance

머신 러닝에 사용하는 데이터는  
값들이 모두 균등하게 나눠져 있다고   
가정한 후 모델에 적합하는 경우가 많습니다.
  
분포를 살피지 않고 모델을 적합할 경우  
빈도가 높은 값들에 과적합(overfitting) 되버립니다.  

즉, __정확도__(Accuracy, 전체 중 정답을 맞춘 정도)는 높은데  
__재현률__(Recall, 실제로 맞는 것을 정답이라 고른 비율)은 낮을 수 밖에 없습니다.

실생활에서는 균형잡힌 데이터보다  
불균형의 데이터들이 넘쳐납니다.  

이를 처리하기 위해선 어떻게 해야할까요?

---

## 불균형 데이터 처리 방법
불균형 데이터를 처리하는 방법은 여러가지가 있습니다.   
그 중 두 가지를 알아보도록 하겠습니다.  


### 1. __오버샘플링(Oversampling)__  

__<무작위 추출>__  
소수의 값을 무작위로 추출하여  
원래 데이터에 더하는 방식  
 
__장점__   
- 데이터의 손실 없이 모두 사용 가능    

__단점__   
- 데이터의 수가 증가 돼 계산량이 많아짐
- 소수 데이터에 과적응 문제가 발생  

__<유의 정보 추출>__   
기준을 정해 해당 하는 값만 추출하여  
원래 데이터에 더하는 방식  


__<합성 데이터 생성>__  
__SMOTE(Synthetic Minority Oversampling Technique)__  
소수의 데이터들을 상호보완하여  
새로운 값을 만들어 추가하는 방법  

`K-NN` 알고리즘을 사용하여  
소수 데이터 주변 값들을 찾고  
그 특성을 반영하도록 데이터를 만든 방법


### 2. __언더샘플링(Undersampling)__  

__<무작위 추출>__  
다수의 값을 무작위로 추출하여  
원 데이터에서 그 값을 삭제하는 방식  

__장점__  
- 크기가 매우 클 때 효과적

__단점__  
- 데이터 정보 손실이 있음

__<유의 정보 추출>__   
유의한 데이터만 남기는 방식  

__EasyEnsemble__  
다수 범주의 집합을 여러개로 나누고  
소수 범주의 집합을 이용하여 

이후 결과를 취합

__BalanceCascade__   
다수 범주의 집합을 N개로 나눔  
그 중 하나를 뽑아  
소수 범주 + 다수 범주(샘플) 로 분류기 하나 만듦  
x가 위의 분류기에 의해 정확하게 분류되면  
x는 충분히 많다고 판단하여 x를 제거  
이 작업을 여러번 반복 해  
마지막으로 하나의 분류기가 만들어지게 함
