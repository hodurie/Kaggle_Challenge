# Clustering
__Clustering__은 비지도학습으로 

학습데이터에 대한 class의 속성들의 정보를 이용하지 않고
데이터의 패턴들을 중심으로 그룹을 분류하는 알고리즘입니다.  

이때, 비슷한 패턴을 가진 개체들을 한 그룹으로 분류하고  
서로 다른 패턴 정보를 가진 개체들은 다른 그룹으로 분류합니다.

즉, 다음과 같은 기준을 이용하여 그룹을 분류합니다. 
1. 군집 내 유사도 최대화
2. 군집 간 유사도 최소화

그룹 간 유사도를 이용하여 군집을 분류 한다면, 
유사도란 어떻게 구해야 할까?

이를 구하는 방식은 거리를 이용하여 구한다. 
거리를 구하는 방식은 
1. 유클리드 거리  
2. 맨하탄 거리
3. 마할로노비스 거리
4. 상관거리

## 유클리드 거리
일반적으로 사용하는 거리로   
두 점의 차를 제곱해 더한 후 루트로 나눠준다.  
이는 직선상의 거리를 의미하며,  
2차원의 거리일 경우 피타고라스의 정리와 동일하다.

## 맨하탄 거리
격자 위에 두 점이 있다고 가정하고,   
한 점에서 다른 점까지 격자를 얼만큼 이동했는지를 나타내는 거리이다.   

## 마할로노비스 거리
변수 내 분산과 공분산을 모두 반영한 거리로,  
통계학적 거리라고도 부른다.  

## 상관 거리
상관 계수를 이용하여 나타낸 거리로,  
0 에서 2 사이의 값을 이용한다. 
0 에 가까울 수록 관측치 간 거리가 가깝다는 얘기며(같은 그룹으로 묶임)
2 에 가까울 수록 관측치 간 거리가 멀다는 얘기이다.(다른 그룹으로 묶임)
