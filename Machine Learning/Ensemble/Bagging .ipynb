# Bagging 
__Bagging__은 __Boostrapping aggregating__의 줄임말로  
__Boostrapping__ 기법으로 생성된 모델들을 결합한 알고리즘입니다.   

__Bagging__을 설명드리기 앞서  
__Boostrapping__에 대해 설명 드리겠습니다.  
> ## __Boostrapping__  
> __Boostrapping__이란   
> 훈련 데이터를 단순 반복 추출하여 원 데이터의 수 만큼 추출하는 방식으로  
> 위와 같은 방식을 여러번 반복해 표본을 만드는 방식입니다.   
>
> 이때, 추출이 진행하면서 새롭게 통계량을 구하하여      
> 기존 모델과 달리 여러 통계량을 기준으로 모수를 추정해
> 모수 분포를 더 안정성 있게 추정하는 기법입니다.  

__Bagging__은  
__Boostrapping__을 이용하여 여러 모델을 만들어  
모수를 추정하는데 여러개의 통계량을 사용하여 모델의 안정성을 높이는 기법입니다.  

따라서, 예측 모형의 변동성이 크다고 예상 될 때 이를 줄이기 위해 사용하며  
분산을 최소화하여 정확도를 향상시키게 됩니다. 

## 모델 생성 방식
1. __Boostrapping__을 이용하여 샘플 추출
2. 각 샘플마다 모델 적합
3. 각 모델들을 결합하여 하나의 모델을 생성
